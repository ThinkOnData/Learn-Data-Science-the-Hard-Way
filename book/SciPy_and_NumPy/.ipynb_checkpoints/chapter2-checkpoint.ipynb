### 2.1 NumPy Arrays

```python
import numpy as np
```

create an array with 10e4 elements

```python
arr = np.arange(10e4)
```

coverting array to list

```python
larr = arr.tolist()
```

what an array can do

```python
def list_times(alist, scalar):
    for i, val in enumerate(alist):
        alist[i] = val * scalar
    return alist
```

Using IPython's magic timeit command

```python
timeit arr**1.1
```

```python
timeit list_times(larr, 1.1)
```

Thus, the ndarray is more effective

## 2.2 Boolean Statements and Numpy Arrays

```python
img1 = np.zeros((20, 20)) + 3
img1
```

```python
img1[4: -4, 4: -4] = 6
```

```python
img1[7: -7, 7: -7] = 9
```

```python
img1
```

```python
index1 = img1 > 2
index2 = img1 < 6
compound_index = index1 & index2
```

```python
compound_index = (img1 > 3) & (img1 < 7)
```

```python
img2 = np.copy(img1)
```

```python
img2[compound_index] = 0
```

```python
index3 = img1 == 9
index4 = (index1 & index2) | index3
img3 = np.copy(img1)
img3[index4] = 0
```
