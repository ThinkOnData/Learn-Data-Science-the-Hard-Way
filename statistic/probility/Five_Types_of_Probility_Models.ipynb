# How To Implement These 5 Powerful Probability Distributions In Python

## 二项分布（Binomial Distribution）

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
n = 10
p = 0.3
k = np.arange(0, 21)
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
plt.plot(k, binomial, 'o-')
plt.title('Binomial: n=%i, p = %.2f' %(n, p), fontsize = 15)
plt.xlabel('Number of successes', fontsize = 15)
plt.ylabel('Probility of successes', fontsize = 15)
plt.show()
```

```python
binom_sim = data = stats.binom.rvs(n = 10, p = 0.3, size = 10000)
print 'Mean: %g' %np.mean(binom_sim)
print 'SD: %g' %np.std(binom_sim, ddof = 1)
```

```python
plt.hist(binom_sim, bins = 10, normed = True)
plt.xlabel('x')
plt.ylabel('density')
plt.show()
```

## 泊松分布（Poisson Distribution）

```python
rate = 2
n = np.arange(0, 10)
y = stats.poisson.pmf(n,rate)
y
```

```python
data = stats.poisson.rvs(mu = 2, loc = 0, size = 10000)
print 'Mean: %g' % np.mean(data)
print 'SD: %g' % np.std(data)
```

```python
plt.hist(data, bins = 9, normed = True)
plt.xlim(0, 10)
plt.xlabel('Number of accodemts')
plt.title('Simulating Poisson Random Variables')
```

## 正态分布（Normal Distribution）

```python
mu = 0
sigma = 1
x = np.arange(-5, 5, 0.1)
x
```

```python
y = stats.norm.pdf(x, 0, 1)
y
```

```python
plt.plot(x, y)
plt.title('Norms: $\mu$=%.1f, $\sigma^2$=%.1f' %(mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability Density')
```

## β分布（Beta Distribution）

```python
a = 0.5
b = 0.5
x = np.arange(0.01, 1, 0.01)
y = stats.beta.pdf(x, a, b)
```

```python
plt.plot(x, y)
plt.title('Beta: a=%.1f, b=%.1f' %(a, b))
plt.xlabel('x')
plt.ylabel('Probability Density')
```

```python
a = 0.1
b = 0.7
x = np.arange(0.01, 1, 0.01)
y = stats.beta.pdf(x, a, b)
plt.plot(x, y)
plt.title('Beta: a=%.1f, b=%.1f' %(a, b))
plt.xlabel('x')
plt.ylabel('Probability Density')
```

## 指数分布（Exponential Distribution）

```python
lambd = 0.5
x = np.arange(0, 15, 0.1)
y = lambd * np.exp(-lambd * x)
plt.plot(x, y)
plt.title('Exponential :$\lambda$=%.2f' %lambd)
plt.xlabel('x')
plt.ylabel('Probability Density')
```

```python
data = stats.expon.rvs(scale = 2, size = 1000)
print 'Mean: %g' % np.mean(data)
print 'SD: %g' % np.std(data, ddof = 1)
```

```python
plt.hist(data, bins = 20, normed = True)
plt.xlim(0, 15)
plt.title('Simulating Exponential Random Variables')
```
